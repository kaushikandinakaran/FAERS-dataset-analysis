```python
import pandas as pd
import matplotlib.pyplot as plt
# pd.set_option('display.max_rows', None)
```

```python
year = 2023
```

```python

# Load the dataset
drug_df = pd.read_csv(f"yearly data/{year}/DRUG_{year}.csv")
```

```python
# List of asthma-related drugs
asthma_drugs = [
    "Albuterol", "Salbutamol", "Levalbuterol", "Xopenex", "Symbicort", "Advair",
    "Fluticasone", "Flovent", "Flonase", "Budesonide", "Pulmicort", "Montelukast",
    "Singulair", "Ipratropium", "Combivent", "Serevent", "Qvar", "DuoNeb", "Theophylline"
]

# Filter the DataFrame for asthma-related records
asthma_related_df = drug_df[drug_df['drugname'].str.contains('|'.join(asthma_drugs), case=False, na=False)]

# Display the shape of the filtered DataFrame and a sample
print(asthma_related_df.shape)
print(asthma_related_df.head())

```
